```python
!pip install numpy matplotlib scikit-learn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)
area = np.random.randint(500, 5000, 100).reshape(-1, 1)  # Property area (sq. ft.)
price = (area * 300) + np.random.randint(20000, 50000, 100).reshape(-1, 1)  # Price in $
```

```python
X_train, X_test, y_train, y_test = train_test_split(area, price, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Intercept: {model.intercept_[0]:,.2f} $")
print(f"Coefficient (Price per Sq. Ft.): {model.coef_[0][0]:,.2f} $")
```

```python
# 5️⃣ Make Predictions on the Test Set
y_pred = model.predict(X_test)

# 6️⃣ Evaluate the Model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:,.2f}")
print(f"R² Score: {r2:.2f}")

# 7️⃣ Visualize the Results
plt.scatter(X_test, y_test, color="blue", label="Actual Prices")
plt.plot(X_test, y_pred, color="red", linewidth=2, label="Predicted Prices")
plt.xlabel("Area (sq. ft.)")
plt.ylabel("Price ($)")
plt.title("Property Price Prediction")
plt.legend()
plt.show()

# 8️⃣ Predict for a New Property
new_area = np.array([[2500]])  # Predict price for a 2500 sq. ft. house
predicted_price = model.predict(new_area)
print(f"Predicted Price for 2500 sq. ft.: {predicted_price[0][0]:,.2f} $")
```
